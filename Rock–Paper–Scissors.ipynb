# 📌 Reflection Questions

---

### ✔ **1) Are the results roughly balanced between options? Why?**

Yes. Because both the player and the computer use uniformly random choices,  
the outcome distribution converges toward one–third for each result (win, lose, draw)  
as the number of rounds increases.  
This behavior is expected due to the fairness and symmetry of the Rock–Paper–Scissors game.

---

### ✔ **2) Why are small simulations not perfectly balanced?**

Small simulations suffer from random fluctuations due to the limited number of rounds.  
The sample size is small, so the percentages can deviate strongly from the expected 33%.  
As the number of rounds increases, the results stabilize due to the  
**Law of Large Numbers**, which makes randomness look more “balanced.”

---

### ✔ **3) How does Python generate random numbers?**

Python uses the **Mersenne Twister** pseudo-random number generator (PRNG).  
It is deterministic, extremely fast, and produces values that are *statistically uniform*.  
Although it is deterministic, it behaves randomly for simulation purposes.

---

### ✔ **4) How can we create a smarter strategy instead of random?**

A smarter strategy can analyze the opponent’s **previous patterns** and predict the next move.
Examples of advanced strategies:

- **Frequency-based prediction:** choose the counter move to the opponent’s most frequent choice  
- **Cycle detection:** if the opponent follows a repeated pattern (R → P → S), detect the cycle  
- **Markov Chain strategy:** predict moves based on transition probabilities  
- **Reinforcement learning:** train an agent to improve based on rewards (advanced)

---

### ✔ **5) What is the difference between Game Logic and the Interface Layer?**

- **Game Logic:**  
  Contains the rules of the game (who wins, random move generation, simulation).  
  It works independently of any interface.

- **Interface Layer (Tkinter):**  
  Only displays results, buttons, text, score…  
  It depends on the logic but does NOT contain any rules.

Separation makes the system more organized, reusable, and easier to debug.
